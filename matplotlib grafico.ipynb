```python
# Carga de módulos de matplotlib
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

# Carga de seaborn
import seaborn as sns

import numpy as np

# Carga de pandas
import pandas as pd
```

```python
data = {
    "año" : ["1945", "1955", "1971", "1987", "1993", "1997", "2005", "2014", "2020"],
    "area_manglar" : [222875.51, 231432.06, 231483.89, 240477.58, 229418.98, 221255.26, 213449.71, 215651.11, 215062.34],
    "area_vegCercana" : [26721.3, 61595.46, 59079.13, 51044.35, 47519.31, 56678.66, 64015.63, 74269.12, 82351.1]
}

areas0 = pd.DataFrame(data)

areas1 = areas0.rename(columns={
    "area_manglar" : "Área del manglar",
    "area_vegCercana" : "Área de la vegetación cercana"
})

areas1
```

```python
# Gráfico de líneas
areas1.plot.line(x="año", 
                        y=["Área del manglar", "Área de la vegetación cercana"],
                        color={"Área del manglar": "green", 
                               "Área de la vegetación cercana": "red"}
                       , linewidth=7.0)

plt.xlabel("Años")
plt.ylabel("metros cuadrados")


plt.savefig("foo.png")
```
