### Esto es para cargar solo una fila con cierto criterio que encuentre, en este caso se guarda en una variable llamada jaguares

Ojo que se carga felinos como el dataframe, luego dentro se le dice que usted quiere comparar el dataframe con la columna que queremos comparar, en este caso species y se busca "Puma concolor" que es el nombre científico que tiene esas filas que andamos buscando.

```python
pumas = felinos[felinos["species"] == "Puma concolor"]
```

```python
pumas
```

```python
pumas_liguistas = felinos[(felinos["species"] == "Puma concolor") & (felinos["stateProvince"] == "Alajuela")]
pumas_liguistas[["stateProvince", "species", "scientificName"]]
```

```python
pumas_guanacastecos = felinos[(felinos["genus"] == "Puma") & (felinos["stateProvince"] == "Guanacaste")]
pumas_guanacastecos
```

```python
pumas_guanacos = felinos[((felinos["species"] == "Puma concolor") | (felinos["species"] == "Puma yagouaroundi")) & (felinos["stateProvince"] == "Guanacaste")]
pumas_guanacos
```

### Acá lo que hice fue lo mismo de arriba pero con una función de pandas, le digo que en la columna de "species" busque una palabra ".str" que contenga ".contains" la palabra "Puma" y la condición de Guana

```python
pumas_guanacos = felinos[(felinos["species"].str.contains("Puma")) & (felinos["stateProvince"] == "Guanacaste")]
pumas_guanacos
```
