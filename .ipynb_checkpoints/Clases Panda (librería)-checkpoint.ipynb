```python
import pandas as pn
primos = [2,3,5,7,11]
serie_primos = pn.Series(primos)

serie_primos
```

# Dataframes

```python
import pandas as pd
# Dataframe construído a partir de dos series

datos = {
  "pais": ["PA", "CR", "NI"],
  "poblacion": [4.1, 5.0, 6.6]
}

paises = pd.DataFrame(datos)

paises
```

El comando .loc[] en las llaves cuadradas poner la posición que quiere que imprima

```python
paises.loc[1]
```

```python
paises.loc[[1,2]]
```
