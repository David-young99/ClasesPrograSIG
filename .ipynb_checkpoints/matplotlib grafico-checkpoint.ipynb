```python
# Carga de módulos de matplotlib
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

# Carga de seaborn
import seaborn as sns

import numpy as np

# Carga de pandas
import pandas as pd
```

```python
data = {
    "año" : ["1945", "1955", "1971", "1987", "1993", "1997", "2005", "2014", "2020"],
    "area_manglar" : [222492, 231576.09, 231627.94, 241310.11, 230510.72, 221255.26, 214718.48, 220309.33, 214049],
    "area_vegCercana" : [26737.97, 61633.85, 59115.96, 51076.20, 46599.99, 56678.66, 62237.06, 70541.87, 82351.10]
}

areas0 = pd.DataFrame(data)

areas1 = areas0.rename(columns={
    "area_manglar" : "Área del manglar",
    "area_vegCercana" : "Área de la vegetación cercana"
})

areas1
```

```python
# Gráfico de líneas
areas1.plot.line(x="año", 
                        y=["Área del manglar", "Área de la vegetación cercana"],
                        color={"Área del manglar": "green", 
                               "Área de la vegetación cercana": "red"}
                       , linewidth=7.0)

plt.title("Cambios en el área del manglar y vegetación cercana")
plt.xlabel("Años")
plt.ylabel("metros cuadrados")


plt.savefig("foo.png")
```
