```python
# Carga de la biblioteca pandas con el alias pd
import pandas as pd

# Carga de geopandas
import geopandas as gpd
```

```python
#version de geopandas
gpd.__version__
```

```python
# Lectura de datos de países de Natural Earth,
# uno de los conjuntos de datos de ejemplo incluídos en geopandas

paises = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Visualización de los datos
paises
```

```python
#Paises con más de esa cantidad de población
paises[paises["pop_est"] > 1000000000]
```

```python
europa = paises.loc[paises["pop_est"] > 10000000, ["name","pop_est"]]

europa
```

```python
#Para contar la extensión de un df
len(europa.index)
```

```python
paises.info()
```

```python
#Práctica

america = paises[(paises["continent"] == "North America") | (paises["continent"] == "South America")]


america
```

```python
len(america.index)
```

```python
america.plot()
```

```python
america.plot(column="pop_est", legend=True)
```

```python
# Mapa de coropletas con paleta de colores y esquema de clasificación
paises.plot(column = "pop_est", 
            legend=True,
            cmap='OrRd', 
            scheme='quantiles',
            figsize=(20, 20)
            )
```

```python
#Para exportarlo es el siguiente método, hay un parámetro que se pone "driver="gpkg", ahó se pone el formato, sino se pone nada se exporta en shp
america.to_file("c:/users/B89124/ClasesPrograSIG/datos/geopandas/america.shp", layer="america")
```
